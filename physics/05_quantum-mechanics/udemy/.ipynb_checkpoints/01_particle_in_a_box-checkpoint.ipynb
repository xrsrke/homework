The particle is in between `x=0` and `x=a`

$dd\psi = \frac{2m}{h^2}(V-E)\psi$





```python
E = 5
psi = 0 # the wave function at the starting position
a = 1

h = 1 # plank constant
m = 1 # mass

dx = a / 1000
dpsi = 1 # the first derivative of psi
```

```python
import matplotlib.pyplot as plt
```

`m` is mass, `h` is Plank constant, `E` is the energy, `psi` is the initial wave function

`dpsi` is the first derivative of the wave function

```python
m, h, E, psi, dpsi
```

```python
a
```

Simulate the wave function in the time-independent Schrödinger equation

**Hints:**
- The time-independent Schrödinger equation can be rearanged into

$\frac{\partial^2}{\partial x^2} \psi(x)=\frac{2 m}{\hbar^2}(V-E) \psi(x)$
- Forward Euler Method: $f(x_{n+1}) =f(x_n)+\Delta x \cdot f'(x_n)$

```python
x = 0
V = 0 # potential energy inside the box
```

```python
x_vals = []
psi_vals = []
```

```python
while x <= a:
    # the second derivative of the wave function
    ddpsi = (2*m/h**2)*(V-E)*psi
    dpsi = dpsi + ddpsi * dx
    psi = psi + dpsi*dx
    
    x += dx
    x_vals.append(x)
    psi_vals.append(psi)
    
    # can be ignored
    plt.plot(x_vals, psi_vals)
    plt.xlabel('x')
    plt.ylabel('$\Psi$ function')
```
