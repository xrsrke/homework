```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
hbar = 1.055e-34  # reduced Planck constant
m = 9.11e-31      # electron mass
L = 1e-9          # width of the well (1 nm)
N = 1000          # number of points in the grid
dx = L / N        # grid spacing
```

```python
# Define the potential function
def V(x):
    if x < 0 or x > L:
        return np.inf  # infinite potential outside the well
    else:
        return 0      # zero potential inside the well
```

```python
# Create the grid
x = np.linspace(0, L, N)
```

```python
# Define the kinetic energy operator
def K(psi):
    dpsi = np.gradient(psi, dx)
    return -(hbar**2 / (2 * m)) * np.gradient(dpsi, dx)
```

```python
# Define the time-independent SchrÃ¶dinger equation
def TISE(psi):
    return K(psi) + V(x) * psi
```

```python
# Set up the initial wave function (infinite potential outside the well)
psi = np.zeros(N)
psi[1:-1] = np.sin(np.pi * x[1:-1] / L)
```

```python
# Use the finite difference method to solve the TISE
max_iter = 1000
tolerance = 1e-6
```

```python
for i in range(max_iter):
    psi_new = psi + TISE(psi) * dx**2
    psi_new /= np.sqrt(np.sum(psi_new**2 * dx))
    if np.sqrt(np.sum((psi_new - psi)**2 * dx)) < tolerance:
        psi = psi_new
        break
    else:
        psi = psi_new
```

```python


# Plot the probability density function
plt.plot(x, psi**2)
plt.xlabel("Position (m)")
plt.ylabel("Probability density")
plt.show()
```
