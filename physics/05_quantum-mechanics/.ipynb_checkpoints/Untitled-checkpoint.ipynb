```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the potential function
def V(x):
    return x ** 2

# Define the wavefunction function and its derivative
def psi(y, x):
    psi, psi_prime = y
    return [psi_prime, 2 * (V(x) - E) * psi]

# Set the boundary conditions
psi_0 = 0
psi_prime_0 = 1

# Define the x range and step size
x = np.linspace(-10, 10, 1000)
h = x[1] - x[0]

# Define the energy range and step size
E_range = np.linspace(0, 10, 1000)
dE = E_range[1] - E_range[0]

# Calculate the potential energy
V_x = V(x)

# Calculate the wavefunction and energy for each energy value
psi_E = []
for E in E_range:
    sol = odeint(psi, [psi_0, psi_prime_0], x)
    psi_x = sol[:, 0]
    psi_E.append(psi_x[-1])
psi_E = np.array(psi_E)

# Normalize the wavefunction
norm = np.sqrt(np.sum(psi_E ** 2 * dE))
psi_E /= norm

# Plot the potential energy and the normalized wavefunction
plt.plot(x, V_x, 'k', label='Potential Energy')
plt.plot(x, psi_E, 'r', label='Wavefunction')
plt.xlabel('x')
plt.ylabel('Energy')
plt.legend()
plt.show()
```
