### 6. Append

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class BaseLinkedList:
    def __init__(self, value = None):
        
        if value == None:
            self.head = None
            self.tail = None
            self.length = 0
        else:
            # create a new node
            new_node = Node(value)
            self.head = new_node
            self.tail = new_node
            self.length = 1
    
    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
```

```python
class LinkedList(BaseLinkedList):
    pass
```

```python
l = LinkedList(); l.__dict__
```

Add a method `append` for add a new item to the end of the list. For each item appended, the length of the list increases by 1

```python
class LinkedList(BaseLinkedList):
    
    def append(self, value):
        new_node = Node(value)
        
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        
        self.length += 1
```

```python
my_list = LinkedList(1) # create a linked list with value 1
my_list.length
```

```python
my_list.append(2); my_list.print_list()
```

```python
my_list.length
```

Add a method `pop` for remove an item at the end of the list. For each item removed, the length of the list decreases by 1

```python
class LinkedList(BaseLinkedList):
    def pop(self):
        if self.length == 0:
            return None
        temp = self.head
        pre = self.head
        while(temp.next is not None):
            pre = temp
            temp = temp.next
        self.tail = pre
        self.tail.next = None
```
