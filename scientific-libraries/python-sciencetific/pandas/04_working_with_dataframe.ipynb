# 4. Working with Dataframe

Points about dataframe:
- Dataframe is two dimensional data
- Each column in dataframe is a series
- Dataframe could be heterogenous (have many dtypes)

### 3. Creating A DataFrame

```python
import pandas as pd
```

```python
names = ['Olga', 'Andrew', 'Nicole']
```

```python
ages = [12, 29, 15]
```

Create a pandas table as bellow

```python
df = pd.DataFrame({'names': names, 'age': ages})
```

```python
df
```

```python
df.shape
```

### 4. Four More Ways to Build DataFrames

##### From series

```python
series_name = pd.Series(names)
```

```python
series_ages = pd.Series(ages)
```

```python
series_name
```

```python
type(series_name), type(series_ages)
```

```python
df2 = pd.DataFrame({'name': series_name, 'ages': series_ages})
```

```python
df2
```

##### From dictionaries

Define a new dictionary `new_dict` and create a dataframe from it

```python
new_dict = {0: 'Olga', 1: 'Andrew', 2: 'Nicole'}
```

```python
df3 = pd.DataFrame({'name': new_dict})
```

```python
df3
```

##### From list of dicts

Create a list of dictionaries `new_dict` and create a dataframe from it

```python
new_dict = [
    {'name': 'Olga', 'age': 12},
    {'name': 'Andrew', 'age': 29},
    {'name': 'Nicole', 'age': 15},   
]
```

```python
df4 = pd.DataFrame(new_dict)
```

```python
df4
```
